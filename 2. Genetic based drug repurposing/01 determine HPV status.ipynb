#### Combine paper and travis case IDs as hpv positive case ids

```python
### paper identified known HPV positive cases and files
paper_hpv_positive_case_ids = pd.read_csv('Data/Supplementary data Nulton/2. HNC HPV Types.csv')['HNSCC Barcode'].tolist()
paper_hpv_all_case_ids = pd.read_csv('Data/Supplementary data Nulton/TCGA CASE ID.csv')['TCGA CASE ID'].tolist()
### convert to only case ids
paper_hpv_all_case_ids = [x.split('-')[0] + '-' + x.split('-')[1] + '-' + x.split('-')[2] for x in paper_hpv_all_case_ids]

print("number of unique case ids from paper for HPV positive cases: ", len(set(paper_hpv_positive_case_ids)))
print("number of unique case ids from paper for all cases: ", len(set(paper_hpv_all_case_ids)))
```

```python
### Somatic mutation data
somatic_mutation = pd.read_csv('Data/TCGA/SOM/cohortMAF.2025-07-29.maf', sep= '\t')
somatic_mutation['Case ID'] = somatic_mutation['Tumor_Sample_Barcode'].str.split('-').str[0:3].str.join('-').tolist()

### filter to nulton SOM cohort for use in downstream analyses
nulton_somatic_cohort = somatic_mutation[somatic_mutation['Case ID'].isin(paper_hpv_all_case_ids)]
nulton_somatic_cohort.to_csv('Data/TCGA/SOM/nulton_somatic_mutation_cohort.csv', index= False)
```

```python
### number of unique cases in the overall paper cohort
print("number of unique case ids from paper for all cases: ", len(set(paper_hpv_all_case_ids)))
### overlap of all case ids and positive case ids from paper
overlap_case_ids = set(paper_hpv_all_case_ids).intersection(set(paper_hpv_positive_case_ids))
print("number of overlapping case ids between all cases and positive cases from paper: ", len(overlap_case_ids))

### number of hpv negatie cases in paper cohort
paper_hpv_negative_case_ids = list(set(paper_hpv_all_case_ids) - set(paper_hpv_positive_case_ids))
print("number of unique case ids from paper for HPV negative cases: ", len(set(paper_hpv_negative_case_ids)))
```

```python
### identify HPV negative cases
paper_hpv_negative_case_ids = list(set(paper_hpv_all_case_ids) - set(paper_hpv_positive_case_ids))
print("number of unique case ids from paper for HPV negative cases: ", len(set(paper_hpv_negative_case_ids)))
```

```python
HPV_POSITIVE_DF = pd.DataFrame()
HPV_POSITIVE_DF['TCGA CASE ID'] = paper_hpv_positive_case_ids
HPV_POSITIVE_DF.to_csv('Results/HPV results/HPV positive patients.csv')

HPV_NEGATIVE_DF = pd.DataFrame()
HPV_NEGATIVE_DF['TCGA CASE ID'] = paper_hpv_negative_case_ids
HPV_NEGATIVE_DF.to_csv('Results/HPV results/HPV negative patients.csv')
```
