```python
# proportion of patients with a death record
def death_proportion():
    """
    proporiton of patients witha  death record
    
    """

    # get the number of unique patients
    patients = PCDM_DEMOGRAPHIC['PATID'].nunique()
    #patients = 5941
    # get the number of unqiue patients with a death record
    death_patients = PCDM_DEATH['PATID'].nunique()

    # Calcul;at ehte rproporiton
    proportion = death_patients/ patients

    return proportion

print(death_proportion())
```
